```python
from ddgs import DDGS
from fastdownload import download_url
from fastai.vision.all import *
from urllib.error import HTTPError
import gradio as gr
import shutil
shutil.rmtree('food', ignore_errors=True)
```

```python
with DDGS() as ddgs:
    results = ddgs.images('pizza food', max_results=150)
    results = list(results)
    print("Images found: " + str(len(results)))
    print(results[0])

for i,r in enumerate(results):
    print(f"Image {i+1}: {r['image']}")
    
Path('food/pizza').mkdir(exist_ok=True, parents=True)

for i in results:
    try:
        download_url(i['image'], 'food/pizza', show_progress=True)
    except (HTTPError,OSError):
        continue
```

## Downloading Images of Burgers and Pizza

```python
with DDGS() as ddgs:
    results = ddgs.images('cheeseburger', max_results=150)
    results = list(results)
    print("Images downloaded: " + str(len(results)))
    print(results[0])

for i,r in enumerate(results):
    print(f"Image {i+1}: {r['image']}")
    
Path('food/burger').mkdir(exist_ok=True, parents=True)

for i in results:
    try:
        download_url(i['image'], 'food/burger', show_progress=True)
    except (HTTPError,OSError):
        continue
```

## Cleanup

```python
bad = verify_images(get_image_files(Path('food')))
for b in bad:
    b.unlink()
print(verify_images(get_image_files(Path('food'))))
```

## Loading Dataset

```python
dls = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=[Resize(192, method='squish')]
).dataloaders(Path('food'), num_workers=0, bs=8)
print("done")

dls.show_batch(max_n=6)
```

## Training

```python
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(3)
```

## Model Performance

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

## Prediction on a new burger image

```python
download_url('https://ontarioporkrecipes.ca/Portals/11/EasyDNNnews/1148/Easy-Smash-Burger-with-Best-Burger-Sauce.jpg', show_progress=True)
learn.predict(Path('Easy-Smash-Burger-with-Best-Burger-Sauce.jpg'))
```

## Export

```python
learn.export('burger-vs-pizza.pkl')
learn_inf = load_learner('burger-vs-pizza.pkl')
```
